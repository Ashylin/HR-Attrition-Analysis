```python
# Let's find all text columns in our dataset
print("=== TEXT COLUMNS ===")
text_columns = df.select_dtypes(include=['object']).columns
print(text_columns.tolist())
print("\nTotal text columns:", len(text_columns))
```

```python
# Convert text columns to numbers
df_model = df.copy()

# Convert Attrition Yes/No to 1/0
df_model['Attrition'] = df_model['Attrition'].map({'Yes': 1, 'No': 0})

# Convert Gender Male/Female to 1/0
df_model['Gender'] = df_model['Gender'].map({'Male': 1, 'Female': 0})

# Convert OverTime Yes/No to 1/0
df_model['OverTime'] = df_model['OverTime'].map({'Yes': 1, 'No': 0})

# Convert remaining text columns automatically
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

text_cols = ['BusinessTravel', 'Department', 'EducationField', 
             'JobRole', 'MaritalStatus']

for col in text_cols:
    df_model[col] = le.fit_transform(df_model[col])

print("All text columns converted to numbers! ✅")
print("\nFirst 5 rows of converted data:")
df_model.head()
```

```python
# Remove columns that are useless for prediction
df_model = df_model.drop(columns=[
    'EmployeeCount',
    'EmployeeNumber', 
    'Over18',
    'StandardHours'
])

print("Useless columns removed! ✅")
print("Remaining columns:", df_model.shape[1])
print("\nAll columns left:")
print(df_model.columns.tolist())
```

```python
# Split data into features and target
X = df_model.drop(columns=['Attrition'])
y = df_model['Attrition']

print("Features (X) shape:", X.shape)
print("Target (y) shape:", y.shape)
print("\nWe are using", X.shape[1], "columns to predict Attrition!")
print("\nTarget value counts:")
print(y.value_counts())
```

```python
# Split data into train and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,      # 20% for testing
    random_state=42,    # makes results reproducible
    stratify=y          # keeps same ratio of Yes/No in both sets
)

print("Data split successfully! ✅")
print("\nTraining set size:", X_train.shape[0], "employees")
print("Testing set size:", X_test.shape[0], "employees")
```

```python
# Build Random Forest Model
from sklearn.ensemble import RandomForestClassifier

# Create the model
model = RandomForestClassifier(
    n_estimators=100,    # 100 decision trees
    random_state=42      # makes results reproducible
)

# Train the model
model.fit(X_train, y_train)

print("Model trained successfully! ✅")
print("Number of trees in our forest:", model.n_estimators)
```

```python
# Test the model
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Make predictions on test data
y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", round(accuracy * 100, 2), "%")

print("\n=== CLASSIFICATION REPORT ===")
print(classification_report(y_test, y_pred))
```

```python
# Chart 8 - Confusion Matrix
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, 
            annot=True, 
            fmt='d', 
            cmap='Blues',
            xticklabels=['Stayed', 'Left'],
            yticklabels=['Stayed', 'Left'])
plt.title('Confusion Matrix', fontsize=16)
plt.xlabel('Predicted', fontsize=12)
plt.ylabel('Actual', fontsize=12)
plt.tight_layout()
plt.savefig('chart8_confusion_matrix.png')
plt.show()
print("Chart 8 saved! ✅")
```

```python
# Chart 9 - Feature Importance
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': model.feature_importances_
}).sort_values('Importance', ascending=False).head(15)

plt.figure(figsize=(10, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance, palette='coolwarm')
plt.title('Top 15 Factors That Cause Employee Attrition', fontsize=16)
plt.xlabel('Importance Score', fontsize=12)
plt.ylabel('Factor', fontsize=12)
plt.tight_layout()
plt.savefig('chart9_feature_importance.png')
plt.show()
print("Chart 9 saved! ✅")
print("\nTop 5 reasons employees leave:")
print(feature_importance.head())
```

```python
# Chart 10 - Overtime vs Attrition
plt.figure(figsize=(8, 5))
sns.countplot(x='OverTime', hue='Attrition', data=df, palette='Set2')
plt.title('Overtime vs Attrition', fontsize=16)
plt.xlabel('OverTime (No = Doesnt work extra, Yes = Works extra hours)', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.tight_layout()
plt.savefig('chart10_overtime_attrition.png')
plt.show()
print("Chart 10 saved! ✅")
```

# 📊 HR Attrition Analysis — Business Insights Report

## 🎯 Project Overview
This project analyzes IBM HR data of 1470 employees to identify 
the key factors driving employee attrition and predict which 
employees are at risk of leaving the company.

## 📈 Key Findings

### 1. Attrition Rate
- Overall attrition rate is **16%** (237 out of 1470 employees left)
- This costs the company significant recruitment and training expenses

### 2. Top Reasons Employees Leave
- **Low Monthly Income** is the #1 factor driving attrition
- **Age** — younger employees (25-35) leave more than older ones
- **Overtime** — employees working overtime leave significantly more
- **Job Satisfaction** — employees with low satisfaction leave more
- **Work Life Balance** — poor balance leads to higher attrition

### 3. Department Insights
- **Sales** department has the highest attrition rate
- **Research & Development** has the most stable workforce

### 4. ML Model Performance
- Random Forest model achieved **85-90% accuracy**
- Model can successfully predict which employees are at risk

## 💡 Business Recommendations
1. **Increase salaries** for low income employees to reduce attrition
2. **Limit overtime** — overworked employees are more likely to quit
3. **Focus on Sales department** — needs immediate retention strategies
4. **Improve job satisfaction** through better work environment
5. **Target younger employees** with career growth opportunities
